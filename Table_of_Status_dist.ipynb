```python
import pandas as pd 
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from matplotlib import style
import tueplots.constants.color.rgb as rgb
from tueplots import bundles
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
#plt.rcParams.update(bundles.neurips2021(usetex=False))
```

```python
data = pd.read_csv('../Data-literacy-project-main/source-data-updated/output_excess-death_poilical-status_hdi.csv')
df = data.drop(['Unnamed: 0'], axis=1)
df = df.replace({'F': 1, 'PF' : 2, 'NF': 3})

sigma = 0.15
df['CL'] = df['CL'].apply(lambda x: np.random.normal(x, sigma))
df['PR'] = df['PR'].apply(lambda x: np.random.normal(x, sigma))

fig, ax = plt.subplots()
ax.scatter(df.PR, df.CL, c=df.Status)
ax.set_ylim(0.5,7.5)
ax.set_xlim(0.5,7.5)
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(MultipleLocator(0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.5))
ax.xaxis.set_tick_params(which='major', color = "white")
ax.yaxis.set_tick_params(which='major', color = "white")
ax.grid(True, which='minor')
ax.invert_yaxis()
ax.xaxis.tick_top()
ax.set_ylabel("Civil liberties", size = 10)
ax.set_title("Political rights", size = 10)

plt.rc('xtick', labelsize=10) #fontsize of the x tick labels
plt.rc('ytick', labelsize=10)

plt.savefig('../Data-literacy-project-main/Plots/table1.pdf',bbox_inches='tight')
```

```python
data = pd.read_csv('../Data-literacy-project-main/source-data-updated/output_excess-death_poilical-status_hdi.csv')
df = data.drop(['Unnamed: 0'], axis=1)
df = df.replace({'F': 1, 'PF' : 2, 'NF': 3})
colors = ["purple", "c","orange"]

fig, ax = plt.subplots()
for i in range(1,4):
    data = pd.DataFrame()
    data = (df
            .groupby(["Status"])
            .get_group(i))
    sigma = 0.15
    data['CL'] = data['CL'].apply(lambda x: np.random.normal(x, sigma))
    data['PR'] = data['PR'].apply(lambda x: np.random.normal(x, sigma))
    ax.scatter(data.PR, data.CL,color=colors[i-1])
    
ax.set_ylim(0.5,7.5)
ax.set_xlim(0.5,7.5)
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(MultipleLocator(0.5))
ax.xaxis.set_minor_locator(MultipleLocator(0.5))
ax.xaxis.set_tick_params(which='major', color = "white")
ax.yaxis.set_tick_params(which='major', color = "white")
ax.grid(True, which='minor')
ax.invert_yaxis()
ax.xaxis.tick_top()
ax.set_ylabel("Civil liberties", size = 10)
ax.set_title("Political rights", size = 10)

plt.rc('xtick', labelsize=10) #fontsize of the x tick labels
plt.rc('ytick', labelsize=10)

plt.savefig('../Data-literacy-project-main/Plots/table1.pdf',bbox_inches='tight')
```
